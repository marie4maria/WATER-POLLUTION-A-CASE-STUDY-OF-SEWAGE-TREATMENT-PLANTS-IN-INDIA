## 2.C) Statistical Hypothesis Testing

<b>2.C.a) Pearsonn Correlation Test between Sewage Generation in Urban Areas and Urban Population</b>

```python
## To Check whether The Sewage Generation in Urban Areas and Population in Urban Areas are Dependent or Independent of each other
## H0 :  Probably The Sewage Generation in Urban Areas and Population are independent
## H1 :  Probably The Sewage Generation in Urban Areas and Population are not independent

from scipy.stats import pearsonr
stat,p=pearsonr(st1['Sewage_Generation_in_Urban_Areas'],st1['Population'])

print('stat=%.3f,p=%.3f' % (stat,p))
if p > 0.05:
    print('Probably The Sewage Generation in Urban Areas and Population are independent')
else:
    print('Probably The Sewage Generation in Urban Areas and Population are not independent')
    
## We Reject the Null hypothesis
```

From the p value we see that we have to reject the Null Hypothesis and Hence our Sewage Generation and Urban population are dependent on each other

<b>2.C.b)Test for Normality for Sewage Generation in Urban Areaas</b>

```python
## To Check whether The Sewage Generation in Urban Areas follow Normality or not.
## H0 :  The Sewage Generation in Urban Areaas Follow Normality
## H1 :  The Sewage Generation in Urban Areaas Do Not Follow Normality

from scipy.stats import shapiro
stat,p=shapiro(st1['Sewage_Generation_in_Urban_Areas'])
print('stat=%.3f,p=%.3f' % (stat,p))

if p>0.05:
    print("The Sewage Generation in Urban Areaas Follow Normality")
else:
    print("The Sewage Generation in Urban Areaas Do Not Follow Normality")
    
    ## We Reject the Null hypothesis
```

As the p value comes less than 0.05 we Reject The Null Hypothesis and The Sewage Generation in Urban Areaas Do Not Follow Normality

<b>2.C.c)Test for Normality for Population in Urban Areas</b>

```python
## To Check whether The Population in Urban Areas follow Normality or not
## H0 :  The Population in Urban Areaas Follow Normality
## H1 :  The Population in Urban Areaas Do Not Follow Normality

from scipy.stats import shapiro
stat,p=shapiro(st1['Population'])
print('stat=%.3f,p=%.3f' % (stat,p))

if p>0.05:
    print("The Population in Urban Areaas Follow Normality")
else:
    print("The Population in Urban Areaas Do Not Follow Normality")
    
    ## We Reject the Null hypothesis
```

# 3) Predictive Analysis

<b> Model 1 </b>

From the Scatter Plot Between Sewage Generation in Urban Areas and The Urban Population we see that there is a High Degree of Linear Relationship existing between the variables.So we can very easily go for a fitting of a Simple Linear Regression Model.

Which can be formulated based on the fact that with the increasing population the amount of waste generated increases.Here in our case study the waste generated is the sewage water that is being discarded in the Urban Areas. For that we Plot the scatter plot between Sewage Generated and Population in the Urban Areas of the states and Union Territories of India.

Population significantly contributes to the Waste Generated and hence the relationship between Population and Sewage generated in the Urban Areas needs to be Studied.

<b> Plotting the Scatter plot between Urban Population and The Sewage Generated in the Urban Areas of the States and Union Territories of India. To depict the degree of relationship between them and fit a suitable regression model.</b>

```python
y=st1["Sewage_Generation_in_Urban_Areas"]
x=st1["Population"]

plt.scatter(x,y)
plt.xlabel('Population in Urban Areas')
plt.ylabel('Sewage Generation in Urban Areas')
plt.title('Fig 10 - Scatter Plot Between Population and Sewage Generated In Urban Areas')
plt.show()
```

From the Scatter plot it is very clear that their exists a strong linear relationship between our variables.
Here Population is our Independent Variable and Sewage Generated is our Dependent Variable.

<b>We Now fit a Simple Linear Regression Model Based on that to predict the values.</b>

```python
import matplotlib.pyplot as plt
from scipy import stats
y=st1["Sewage_Generation_in_Urban_Areas"]
x=st1["Population"]
slope, intercept, r, p, std_err = stats.linregress(x,y)
res = stats.linregress(x,y)

def myfunc(x):
    return slope * x + intercept
mymodel = list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.xlabel('Population In Urban Areas')
plt.ylabel('Sewage Generation In Urban Area')
plt.title('Fig 11 - Fitted Regression Model of Sewage Generation Based Upon Population in Urban Areas')
plt.show()
print("Regression result:", res)
mymodel_SLR1 = list(map(myfunc, [100000]))
print("Predicted value", mymodel_SLR1)
import statsmodels.api as sm
lm1 = sm.OLS(y,x).fit()
lm1.summary()
```

Our Fitted Simple Linear Regression Model Based on Population and Sewage Generation is given by

<b>fitted_Sewage_generation = 0.00015628820181559142 * Population  +  257.28583063133374</b>

<b>Model Interpretations</b>

The slope of the Model or the B1 coefficient is 0.00015628820181559142 which indicates that with every one unit increase in population the sewgae generated will be increasing by an amount of 0.00015628820181559142 tonnes.

The intercept of the Model or the B0 coefficient is 257.28583063133374 which indicates that when the value of Population is zero the average value of the sewage generated is 257.28583063133374 tonnes.

According to our Model for a Place with population of 100000 there is a Sewage Generation of 272.9146508128929 tonnes.

The R-Squarred value of our data is 0.839 which indicates that the independent variable explains 83.9% of the Variability in the dependent Variable. The value of 0.839 indicates that our Model is a good fit and we can move forward with our predictions with this Model.

<b>Model 2</b>

We saw that the Amount of Sewage Generated in the Urban Areas is much more than the Operational Capacity of the Sewage treatment plants. The Operational Capacity of the Sewage Treatment Plants must be at par with the Sewage generation in the Urban Areas. So we need to know what is the extent of the relationship between the Sewage Generated and the Operational Capacity. We model this based on the assumption that Operational Capacity is the value which indicates the actual possible sewage that is getting processed.

<b>Plotting the Scatter plot between The Sewage Generated and The Operational Capacity of the Sewage Treatment Plants in the Urban Areas of the States and Union Territories of India. To depict the degree of relationship between them and fit a suitable regression model.</b>

```python
a=st1["Sewage_Generation_in_Urban_Areas"]
b=st1["Operational_Capacity"]

plt.scatter(a,b)
plt.ylabel('Operational Capacity ')
plt.xlabel('Sewage Generation in Urban Areas')
plt.title('Fig 12 - Scatter Plot Between Sewage Generated In Urban Areas and Operational Capacity of Sewage Treatment Plants')
plt.show()
```

From the Scatter plot we see that there is a moderate degree of relationship between the variables and the relationship can be taken to be Linear. So we fit a simple Linear Regression based on the given datas, under the assumption that more the Sewage Generated in the Urban Areas more is the requirement of the Sewage Treatment Plants. But with Operatioal Capacity based on the Sewage generated we may understand how much of the Sewage Generated is actually being Processed.

<b>We Now fit a Simple Linear Regression Model Based on that to predict the values.</b>

```python
import matplotlib.pyplot as plt
from scipy import stats
a=st1["Sewage_Generation_in_Urban_Areas"]
b=st1["Operational_Capacity"]
slope, intercept, r, p, std_err = stats.linregress(a,b)
res = stats.linregress(a,b)

def myfunc(a):
    return slope * a + intercept
mymodel = list(map(myfunc,a))

plt.scatter(a,b)
plt.plot(a,mymodel)
plt.ylabel('Operational Capacity')
plt.xlabel('Sewage Generation In Urban Area')
plt.title('Fig 13 - Fitted Regression Model of Operational Capacity of STP based upon Sewage Generation in Urban Areas')
plt.show()
print("Regression result:", res)
mymodel_SLR2 = list(map(myfunc, [1000]))
print("Predicted value", mymodel_SLR2)
import statsmodels.api as sm
lm2 = sm.OLS(b,a).fit()
lm2.summary()
```

Our Fitted Simple Linear Regression Model is Given By

<b> fitted_Operational_Capacity = -137.01692858502042 + 0.38444823770034114 * Sewage Generated

<b>Model Interpretations</b>

The slope of the Model or the B1 coefficient is 0.38444823770034114 which indicates that with every one unit increase in sewgae generated the operational capacity will be increasing by an amount of 0.38444823770034114 tonnes.

The intercept of the Model or the B0 coefficient is -137.01692858502042 which indicates that when the value of sewage generated on the average value of Operational Capacity  is -137.01692858502042 tonnes.

According to our Model for every Sewage Generation of 1000 tonnes the Operational Capacity is 247.43130911532074 tonnes.

The R-Squarred value of our data is 0.751 which indicates that the independent variable explains 75.1% of the variability in the dependent Variable. The value of 0.751 indicates that our Model is a good fit and we can move forward with our predictions with this Model.
